```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/NHL_game_log_all.csv')
```

```python
df.head()
```

```python
# Get Season for which game was played
df['Season'] = df.Date.apply(lambda x: f'{x[0:4]}-{str(int(x[0:4])+1)}' if int(x[5:7])>=7 else f'{str(int(x[0:4])-1)}-{x[0:4]}')
```

```python
 # Fix team names where team named switched but same franchise
df['Tm'] = df.Tm.apply(lambda x: 'ANA' if x == 'MDA' else x)
df['Tm'] = df.Tm.apply(lambda x: 'CHI' if x == 'CBH' else x)
df['Tm'] = df.Tm.apply(lambda x: 'ARI' if x == 'PHX' else x)
```

```python
df['Opp'] = df.Opp.apply(lambda x: 'ANA' if x == 'MDA' else x)
df['Opp'] = df.Opp.apply(lambda x: 'CHI' if x == 'CBH' else x)
df['Opp'] = df.Opp.apply(lambda x: 'ARI' if x == 'PHX' else x)
```

```python
# Find Season number and career game number for each line

def player_season(row):
    global player_dict
    return player_dict[row.Player][row.Season]

player_game_count = dict()
def player_num_games(row):
    global player_game_count
    if row.Player in player_game_count:
        player_game_count[row.Player] += 1
    else:
        player_game_count[row.Player] = 1
    return player_game_count[row.Player]

player_dict = dict()
for player in df.Player.unique():
    print(f'\r{player}                           ',end='')
        
    player_dict[player] = dict()
    for season_num,season in enumerate(df[df.Player == player].Season.unique()):
        player_dict[player][season] = season_num+1
    
```

```python
df['Player_Season_Num'] = df.apply(player_season, axis = 1)
```

```python
df['Career_Game_Num'] = df.apply(player_num_games, axis = 1)
```
